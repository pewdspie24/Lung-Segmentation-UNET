```
import torch
from torch import nn
from torch.utils.data import DataLoader, SubsetRandomSampler
import os
import shutil
from dataset import CustomDataset
from model import UNet
# from tqdm import tqdm #for pc
import tqdm.notebook as tq #for colab
from diceCoeff import dice_coeff
import torch.optim as optim
import argparse
from sklearn.model_selection import KFold
import time
from matplotlib import pyplot as plt
```

```
batch_size = 16
LR = 3e-4
epochs = 6 #10
k_folds = 5
torch.manual_seed(24)
kfold = KFold(n_splits=5, shuffle=True)
device = 'cuda' if torch.cuda.is_available() else 'cpu'
dataset = CustomDataset('./lung-seg')

# trainDataset = CustomDataset('./lung-seg/train/')
# trainLoader = DataLoader(trainDataset, batch_size=batch_size, shuffle=True, num_workers=2)
 
# validDataset = CustomDataset('./final/validate/')
# validLoader = DataLoader(validDataset, batch_size=batch_size, num_workers=2)

model = UNet(3,1).to(device)

optimizer = optim.Adam(model.parameters(), lr=LR, weight_decay=0)
criterion = nn.BCEWithLogitsLoss()
#train
losses_train = []
losses_valid = []
max_loss = 100

for fold,(train_idx,val_idx) in enumerate(kfold.split(dataset)):
  train_subsampler = SubsetRandomSampler(train_idx)
  val_subsampler = SubsetRandomSampler(val_idx)
  trainLoader = DataLoader(dataset, batch_size=batch_size, sampler=train_subsampler, num_workers=2)
  valLoader = DataLoader(dataset, batch_size=batch_size, sampler=val_subsampler, num_workers=2)
  # print(train_idx)
  # print(val_idx)
  for epoch in range(epochs):
      running_train_loss = 0.0
      running_valid_loss = 0.0
      model.train()
      for (input, mask) in tq.tqdm(trainLoader): #for colab
      #for (input, mask) in tqdm(trainLoader): #for pc
          optimizer.zero_grad()
          input = input.to(device)
          mask = mask.to(device)
          output = model(input)
          loss = criterion(output, mask)
          loss.backward()
          optimizer.step()
          running_train_loss += loss.item() * input.size(0) 
      epoch_train_loss = running_train_loss / len(trainLoader)
      losses_train.append(epoch_train_loss)
      print("")
      print('Training, Fold {} - Epoch {} - Loss {}'.format(fold+1, epoch+1, epoch_train_loss))

      model.eval()
      with torch.no_grad():
        for (input, mask) in tq.tqdm(valLoader): #for colab
        #for (input, mask) in tqdm(valLoader): #for pc
          input = input.to(device)
          mask = mask.to(device)
          output = model(input)
          loss = criterion(output, mask)
          running_valid_loss += loss.item() * input.size(0)
        epoch_valid_loss = running_valid_loss / len(valLoader)
        losses_valid.append(epoch_valid_loss)
        print("")
        print('Validating, Fold {} - Epoch {} - Loss {}'.format(fold+1, epoch+1, epoch_valid_loss))
  if epoch_valid_loss < max_loss:
      max_loss = epoch_valid_loss
      torch.save(model.state_dict(), "/content/drive/MyDrive/AI-ML/Lane_Segment_PyTorch/checkpointLungSegmentSigBest.pth")
plt.plot(losses_train, label="train")
plt.plot(losses_valid, label="valid") 
plt.legend(loc="upper left")
torch.save(model.state_dict(), "/content/drive/MyDrive/AI-ML/Lane_Segment_PyTorch/checkpointLungSegmentSigLast.pth")
```
